**import libraries**

**Load the dataset and create our dataframe**

```python
# Load the iris dataset
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head() 
```

**here we will use the kmeans algorithme because we have unsupervised problem (Clustering) and the elbow methode to return the optimum number of clusters in this dataset**

```python
'''Elbow Methode'''
iris_df = iris_df.iloc[:, [0, 1, 2, 3]].values
from sklearn.cluster import KMeans
sum_of_squared_distances=[]
K=range(2,10)
for i in K:
    Km=KMeans(n_clusters=i)
    Km=Km.fit(iris_df)#train du kmeans
    sum_of_squared_distances.append(Km.inertia_)
    


 
```

**plotting the result**

```python
plt.figure()
plt.plot(K,sum_of_squared_distances,'b*-')
plt.xlabel('K')
plt.ylabel('sum_of_squared_distances')
plt.title('Elbow ethod For Optimal K')
```

**Although we already know the answer is 3 as there are 3 unique class in Iris flowers**

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(iris_df)
Label=kmeans.labels_
centers=kmeans.cluster_centers_
```

**we can use another methode of clustering to indicate that the k number of clusters that we working with is the best choice**

we can use silhouette method: **Silhouette analysis** can be used to study the separation distance between the resulting clusters. The silhouette plot displays a measure of how close each point in one cluster is to points in the neighboring clusters and thus provides a way to assess parameters like number of clusters visually.

```python
'''Coefficient of silhouette'''
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import SilhouetteVisualizer
```

```python
K=range(3,10)
for i in K:
    Km=KMeans(n_clusters=i)
    Km=Km.fit(iris_df)#train du kmeans
    labels=Km.labels_
    centers=Km.cluster_centers_
    score=silhouette_score(iris_df,labels)
   
    #print(i,":",score)
    plt.figure(figsize=(15, 10))
    plt.subplot(1,2,2)
    plt.scatter(iris_df[:,0],iris_df[:,1],c=labels,cmap="plasma") #c=labels :each cluster with each color coulour
    plt.scatter(centers[:,0],centers[:,1],c='green',s=80)
    #cmap=plasma==>use customs colors
    plt.subplot(1,2,1)
    model = SilhouetteVisualizer(Km)
    model.fit(iris_df)
    model.show()
```

```python
#silhouette avg 
k=range(3,10)
for i in k:
    cluster=KMeans(n_clusters=i)
    cluster_labels=cluster.fit_predict(iris_df)
    silhouette_avg=silhouette_score(iris_df,cluster_labels)
    print("For n_clusters=",i,"the average silhouette_score is :",silhouette_avg)
```

**As we can see here in the result when n_cluster=3 we have the high silhouette_score**
